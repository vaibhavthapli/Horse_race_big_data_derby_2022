# race_id is connect both dataset

```python
#merge
new_data = pd.merge(rrh, rrr, on='race_id', how ="inner")
```

```python
new_data.isnull().sum()
```

```python
# Convert 'race_date' to datetime
new_data['race_date'] = pd.to_datetime(new_data['race_date'])
```

```python
def determine_hk_season(date):
    month = date.month
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:  # [9, 10, 11]
        return 'Autumn'

new_data['Season'] = new_data['race_date'].apply(determine_hk_season)
```
