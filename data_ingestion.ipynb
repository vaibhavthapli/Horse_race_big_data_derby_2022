```python
import pandas as pd
```

```python
#merge
new_data = pd.merge(df, df_r, on='race_id', how ="inner")
```

```python
new_data.shape
```

```python
new_data.head()
```

```python
# Count occurrences of each horse_name
value_counts = new_data.horse_name.value_counts()

horse_name_count1 = (value_counts == 1).sum()
print(f"Number of horse_name with exactly 1 and occurrences: {horse_name_count1}")
horse_name_count = (value_counts == 2).sum()
print(f"Number of horse_name with exactly 2 and occurrences: {horse_name_count}")
```

```python
# Step 1: Count occurrences of each horse_name
value_counts = new_data['horse_name'].value_counts()

# Step 2: Get horse_name values that occur less than or equal to 2 times
horse_names_with_less_than_or_equal_two = value_counts[value_counts <= 2].index

# Step 3: Filter the original DataFrame
filtered_data = new_data[new_data['horse_name'].isin(horse_names_with_less_than_or_equal_two)]

# Display or save the new DataFrame
filtered_data.head(10)
```

```python
filtered_data.shape
```

```python
# we have 391 datapoints where horse race only 1's or twice in his career
```

```python
# jockey
```

```python
# Count occurrences of each jockey
value_counts = new_data.jockey.value_counts()

jockey_name_count1 = (value_counts == 1).sum()
print(f"Number of jockey_name with exactly 1 and occurrences: {jockey_name_count1}")
jockey_name_count2 = (value_counts == 2).sum()
print(f"Number of jockey_name with exactly 2 and occurrences: {jockey_name_count2}")
```

```python
# Step 1: Count occurrences of each jockey
jockey_value_counts = new_data['jockey'].value_counts()

# Step 2: Get jockey values that occur less than or equal to 2 times
jockeys_with_less_than_or_equal_two = jockey_value_counts[jockey_value_counts <= 2].index

# Step 3: Filter the original DataFrame to create a new DataFrame
filtered_jockey_data = new_data[new_data['jockey'].isin(jockeys_with_less_than_or_equal_two)]

# Display or save the new DataFrame
print("Filtered DataFrame for jockey:")
filtered_jockey_data.head()
```

```python
filtered_jockey_data.shape
```

```python


# Merge the two filtered DataFrames (same columns)
merged_filtered_data = pd.concat([filtered_jockey_data, filtered_data], ignore_index=True)

# Remove duplicate rows if needed
merged_filtered_data = merged_filtered_data.drop_duplicates()

# Display or save the merged DataFrame
print("Merged DataFrame:")
merged_filtered_data.shape
```

```python
new_data.info()
```

```python
# Replace "1 DH" to "1", "2 DH" to "2", etc.
new_data['finishing_position'] = new_data['finishing_position'].replace({
    "1 DH": "1", "2 DH": "2", "3 DH": "3", "4 DH": "4",
    "5 DH": "5", "6 DH": "6", "7 DH": "7", "8 DH": "8",
    "9 DH": "9", "10 DH": "10", "11 DH": "11", "12 DH": "12"
})
```

```python
new_data.replace("---","NaN",inplace=True)
new_data.replace("-","NaN",inplace=True)
```

```python
#check missing values again
missing_data_check = new_data.isin(['NaN']) | new_data.isnull()

# Summarize missing values
missing_values_summary = missing_data_check.sum()
```

```python
missing_values_summary
```

```python
# Drop rows in column where value is not a number
new_data = new_data[new_data['finishing_position'].isin([str(i) for i in range(1, 13)])].reset_index(drop=True)
```

```python
missing_values_summary
```

```python
from sklearn.utils import shuffle
# Shuffle the dataset
shuffled_df = shuffle(new_data).reset_index(drop=True)
```

```python
shuffled_df.head()
```

```python
#train_test split
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

# Define the stratified shuffle split
split = StratifiedShuffleSplit(n_splits=2, test_size=0.2, random_state=42)

# Perform the split
for train_index, test_index in split.split(shuffled_df, shuffled_df['finishing_position']):
    train_stratified = shuffled_df.iloc[train_index]
    test_stratified = shuffled_df.iloc[test_index]

# Check the sizes of the splits
print("Stratified Split:")
print(f"Training Size: {len(train_stratified)}, Testing Size: {len(test_stratified)}")

# Verify the distribution of the target variable in each set
print("\nOriginal Distribution:")
print(shuffled_df['finishing_position'].value_counts())

print("\nTraining Set Distribution:")
print(train_stratified['finishing_position'].value_counts())

print("\nTesting Set Distribution:")
print(test_stratified['finishing_position'].value_counts())
```

```python
train_stratified.shape
```

```python
merged_filtered_data.shape
```

Add merged_filtered_data with train data

```python
train_df = pd.concat([train_stratified, merged_filtered_data], ignore_index=True)
# Remove duplicate rows if needed
train_data = train_df.drop_duplicates()
train_data.shape
```

```python
train_data.to_csv('./data/train.csv', index=False)
test_stratified.to_csv('./data/test.csv', index=False)
```
